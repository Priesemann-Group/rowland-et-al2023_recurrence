```python
## Hit vs Miss
num_fits = 10 # Should be 1000, but 40 gives already good results 
n_factors_list = [0,5]
num_bootstrap = 0
activity_type_list = ['shared', 'residual']
parameter_names = ('n_fact', 'session', 'activity_type')
parameter_values = (n_factors_list, list(session_dict.keys()), activity_type_list)

time_begin = time.time()

### Define LFA

session_dict_delayed={}
for session_name,session in session_dict.items():
    session_dict_delayed[session_name] = dask.delayed(session.sel(area=1))

    
lfa_func = dask.delayed(re.factor_analysis.split_shared_residual_part, nout=2)
nested_session_dict = {}
for n_fact in n_factors_list:
    nested_session_dict[n_fact] = {key: {} for key in activity_type_list}
    for session_name,session in session_dict_delayed.items():
        if n_fact == 0:
            nested_session_dict[n_fact]['shared'][session_name] = session
            nested_session_dict[n_fact]['residual'][session_name] = session
            continue
        session_shared, session_residual = lfa_func(session, n_fact=n_fact, fa_type='pca')
        nested_session_dict[n_fact]['shared'][session_name] = session_shared
        nested_session_dict[n_fact]['residual'][session_name] = session_residual

### Define recurrency calculation        
delayed_dict = {}
#future_dict = {}
recurrency_func = dask.delayed(re.variance_cc.prepost_var_cc_and_taupost_of_session_TestOnly_Hit)
for params_to_calc in itertools.product(*parameter_values):
    n_fact, session, activity_type = params_to_calc
    delayed_task = recurrency_func(nested_session_dict[n_fact][activity_type][session], num_bootstrap, num_fits=num_fits)
    #future = client.submit(re.var_cc_of_session, client.compute(nested_session_dict[n_fact][activity_type][session]), num_bootstrap, num_fits=10)
    delayed_dict[params_to_calc] = delayed_task

    
### Begin calculatation

future_dict = client.compute(delayed_dict)
#dask.distributed.progress(future_dict)
```

```python
from scipy.optimize import curve_fit
```

```python
def fit_func(t, A, tau):
    return A*np.exp(-t/tau)
```

```python
n_sessions = len(session_dict)

fig, ax = plt.subplots(n_sessions, 1, figsize=(5,25), dpi=300)

HitAndMiss_tau_hats = np.zeros(n_sessions)    

i_session = -1

for session_name, session in session_dict.items():
    i_session += 1
    targeted_neurons = np.any(session.is_target, axis=1)

    n_neurons = session.shape[0]
    n_frames = session.shape[2]

    pre_frames = session.where(session.frame_type.isin((-1)), drop=True).frame_num.values
    post_frames = session.where(session.frame_type.isin((1)), drop=True).frame_num.values

    testStim_trials = session.where(session.stim_type.isin((1)), drop=True).trial_num.values

    hitTrials = session.where(session.trial_type.isin(("hit")), drop=True).trial_num.values
    missTrials = session.where(session.trial_type.isin(("miss")), drop=True).trial_num.values
    HitAndMiss_trials = session.where(session.trial_type.isin(("hit", "miss")), drop=True).trial_num.values

    testHitTrials = np.intersect1d(testStim_trials, hitTrials)
    testMissTrials = np.intersect1d(testStim_trials, missTrials)
    testHitAndMissTrials = np.union1d(testHitTrials,testMissTrials)

    testHitAndMiss_TargetNeuronAverages = np.zeros((n_neurons,n_frames))
    testHitAndMiss_TargetNeuronAverages[:] = np.nan


    for i_neuron in np.arange(n_neurons)[targeted_neurons]:
        testHitAndMiss_TargetNeuronAverages[i_neuron] = np.nanmean(session[i_neuron][testHitAndMissTrials[session.is_target[i_neuron,testHitAndMissTrials].values]], axis=0)

    testHitAndMiss_delta = np.nanmean(testHitAndMiss_TargetNeuronAverages, axis=0) - np.nanmean(np.nanmean(testHitAndMiss_TargetNeuronAverages, axis=0)[0:pre_frames[-1]])

    ax[i_session].plot(testHitAndMiss_delta, color='red', label='HitAndMiss delta')   

    t = post_frames

    testHitAndMiss_popt, _ = curve_fit(fit_func, t, 
                                testHitAndMiss_delta[post_frames], 
                                p0=(1, 100),
                                maxfev = 10000)

    HitAndMiss_tau_hats[i_session] = testHitAndMiss_popt[1]

    ax[i_session].plot(t, fit_func(t, *testHitAndMiss_popt),
                       color='red', linestyle='--', 
                       label='HitAndMiss delta fit, tau={:.0f}'.format(testHitAndMiss_popt[1]))   

    ax[i_session].axvline(pre_frames[-2]+500/30, color='black')
    ax[i_session].axvline(post_frames[0], color='black', linestyle='--')

    #ax[i_session].set_xlim(session.time[0],session.time[-1])
    ax[i_session].set_ylim(-1,1)

    ax[i_session].set_ylabel('average dF/F')

    ax[i_session].text(x=275, y=0.35, s=session_name, fontsize=8)

    ax[i_session].legend(loc='upper left', fontsize=8)
    
    if session_name == 'Mouse RL116, run 33':
        ax[i_session].set_xlabel('frames')
        plt.show()
```

```python
n_sessions = len(session_dict)

fig, ax = plt.subplots(n_sessions, 1, figsize=(5,25), dpi=300)

Hit_tau_hats = np.zeros(n_sessions)    
Miss_tau_hats = np.zeros(n_sessions) 

i_session = -1

for session_name, session in session_dict.items():
    i_session += 1
    targeted_neurons = np.any(session.is_target, axis=1)

    n_neurons = session.shape[0]
    n_frames = session.shape[2]

    pre_frames = session.where(session.frame_type.isin((-1)), drop=True).frame_num.values
    post_frames = session.where(session.frame_type.isin((1)), drop=True).frame_num.values

    testStim_trials = session.where(session.stim_type.isin((1)), drop=True).trial_num.values

    hitTrials = session.where(session.trial_type.isin(("hit")), drop=True).trial_num.values
    missTrials = session.where(session.trial_type.isin(("miss")), drop=True).trial_num.values
    HitAndMiss_trials = session.where(session.trial_type.isin(("hit", "miss")), drop=True).trial_num.values

    testHitTrials = np.intersect1d(testStim_trials, hitTrials)
    testMissTrials = np.intersect1d(testStim_trials, missTrials)
    testHitAndMissTrials = np.union1d(testHitTrials,testMissTrials)

    testHit_TargetNeuronAverages = np.zeros((n_neurons,n_frames))
    testHit_TargetNeuronAverages[:] = np.nan

    
    testMiss_TargetNeuronAverages = np.zeros((n_neurons,n_frames))
    testMiss_TargetNeuronAverages[:] = np.nan

    for i_neuron in np.arange(n_neurons)[targeted_neurons]:
        testHit_TargetNeuronAverages[i_neuron] = np.nanmean(session[i_neuron][testHitTrials[session.is_target[i_neuron,testHitTrials].values]], axis=0)
        testMiss_TargetNeuronAverages[i_neuron] = np.nanmean(session[i_neuron][testMissTrials[session.is_target[i_neuron,testMissTrials].values]], axis=0)

    testHit_delta = np.nanmean(testHit_TargetNeuronAverages, axis=0) - np.nanmean(np.nanmean(testHit_TargetNeuronAverages, axis=0)[0:pre_frames[-1]])
    testMiss_delta = np.nanmean(testMiss_TargetNeuronAverages, axis=0) - np.nanmean(np.nanmean(testMiss_TargetNeuronAverages, axis=0)[0:pre_frames[-1]])

    ax[i_session].plot(testHit_delta, color='green', label='Hit delta')  
    ax[i_session].plot(testMiss_delta, color='magenta', label='Miss delta')

    t = post_frames

    testHit_popt, _ = curve_fit(fit_func, t, 
                                testHit_delta[post_frames], 
                                p0=(1, 100),
                                maxfev = 10000)

    Hit_tau_hats[i_session] = testHit_popt[1]
    
    testMiss_popt, _ = curve_fit(fit_func, t, 
                                testMiss_delta[post_frames], 
                                p0=(1, 100),
                                maxfev = 10000)

    Miss_tau_hats[i_session] = testMiss_popt[1]

    ax[i_session].plot(t, fit_func(t, *testHit_popt),
                       color='green', linestyle='--', 
                       label='Hit delta fit, tau={:.0f}'.format(testHit_popt[1]))   
    
    ax[i_session].plot(t, fit_func(t, *testMiss_popt),
                       color='magenta', linestyle='--', 
                       label='Miss delta fit, tau={:.0f}'.format(testMiss_popt[1]))   

    ax[i_session].axvline(pre_frames[-2]+500/30, color='black')
    ax[i_session].axvline(post_frames[0], color='black', linestyle='--')

    #ax[i_session].set_xlim(session.time[0],session.time[-1])
    ax[i_session].set_ylim(-1,1)

    ax[i_session].set_ylabel('average dF/F')

    ax[i_session].text(x=275, y=0.35, s=session_name, fontsize=8)

    ax[i_session].legend(loc='upper left', fontsize=8)
    
    if i_session == n_sessions-1:
        ax[i_session].set_xlabel('frames')
        
plt.show()
```
