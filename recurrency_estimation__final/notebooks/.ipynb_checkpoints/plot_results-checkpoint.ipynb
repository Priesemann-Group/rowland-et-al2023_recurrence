```python
## R(Hit and Miss) overview
res_dataset = xr.load_dataarray("ML_nfac0+5_HitAndMissResults_1000fits.nc")
```

```python
var_cc = res_dataset.sel(n_fact=5, activity_type="residual", sample=0, trial_type="HitAndMiss", variable='var_cc').values
mean_var = res_dataset.sel(n_fact=5, activity_type="residual", sample=0, trial_type="HitAndMiss", variable='mean_var').values
```

```python
print(var_cc)
```

```python
def calc_R(mu_V, sigma_CC, N=100000):
    s = sigma_CC / mu_V
    
    return np.sqrt(1-np.sqrt(1/(1+N * s**2)))
```

```python
R = calc_R(mean_var, var_cc, N=50000)
```

```python
print(R)
```

```python
def ML_R_PlotSketch(smin=0, smax=0.1, ax=False):
    f_R = lambda s, N: np.sqrt(1-np.sqrt(1/(1+N * s**2)))

    s_range = np.linspace(smin, smax, 1000)
    
    if not ax:
        fig, ax = plt.subplots(figsize=(2,1.8), dpi=300)

    ax.plot(s_range, f_R(s_range, 100000), color='black')
    
    ax.set_xlim(0,0.1)
    ax.set_ylim(0,1)
    ax.set_yticks([0,0.5,1])
    ax.set_xticks([0,0.05,0.1])
    #ax.axhline(1, color='black', linestyle='--')
    
    ax.set_xlabel(r"$\sigma_{CC}$" + " [norm.]")
    ax.set_ylabel("recurrence " + r"$R$")
```
