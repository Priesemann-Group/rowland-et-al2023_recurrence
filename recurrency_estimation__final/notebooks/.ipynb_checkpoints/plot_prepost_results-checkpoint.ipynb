```python
## R(Hit and Miss) overview
res_S1_dataset = xr.load_dataarray("ML_nfac0+5_HitAndMissResults_1000fits.nc")
res_S2_dataset = xr.load_dataarray("ML_nfac0+5_HitAndMiss_S2Results_1000fits.nc")
```

```python
var_cc = res_dataset.sel(n_fact=5, activity_type="residual", sample=0, trial_type="HitAndMiss", variable='var_cc').values
mean_var = res_dataset.sel(n_fact=5, activity_type="residual", sample=0, trial_type="HitAndMiss", variable='mean_var').values
s = var_cc / mean_var
```

```python
var_cc_S1 = res_S1_dataset.sel(n_fact=5, activity_type="residual", sample=0, trial_type="HitAndMiss", variable='var_cc').values
mean_var_S1 = res_S1_dataset.sel(n_fact=5, activity_type="residual", sample=0, trial_type="HitAndMiss", variable='mean_var').values
s_S1 = var_cc_S1 / mean_var_S1

var_cc_S2 = res_S2_dataset.sel(n_fact=5, activity_type="residual", sample=0, trial_type="HitAndMiss", variable='var_cc').values
mean_var_S2 = res_S2_dataset.sel(n_fact=5, activity_type="residual", sample=0, trial_type="HitAndMiss", variable='mean_var').values
s_S2 = var_cc_S2 / mean_var_S2
```

```python
def calc_R(mu_V, sigma_CC, N=50000):
    s = sigma_CC / mu_V
    
    return np.sqrt(1-np.sqrt(1/(1+N * s**2)))
```

```python
R = calc_R(mean_var, var_cc, N=50000)
```

```python
R_S1 = calc_R(mean_var_S1, var_cc_S1, N=50000)
R_S2 = calc_R(mean_var_S2, var_cc_S2, N=6000)
```

```python
print(R)
```

```python
def ML_R_PlotSketch(smin=0, smax=0.15, ax=False, N=10000, color='black', linestyle='-'):
    f_R = lambda s, N: np.sqrt(1-np.sqrt(1/(1+N * s**2)))

    s_range = np.linspace(smin, smax, 1000)
    
    if not ax:
        fig, ax = plt.subplots(figsize=(2,1.8), dpi=300)

    ax.plot(s_range, f_R(s_range, N), color=color, linestyle=linestyle)
    
    ax.set_xlim(0,0.15)
    ax.set_ylim(0,1)
    ax.set_yticks([0,0.5,1])
    ax.set_xticks([0,0.075,0.15])
    #ax.axhline(1, color='black', linestyle='--')
    
    ax.set_xlabel(r"$\sigma_{CC}$" + " (normalised)")
    ax.set_ylabel("recurrence " + r"$R$")
```

```python
fig, ax = plt.subplots(figsize=(1.25,1.25), dpi=300)

mean_s = np.mean(s)
mean_R = np.mean(R)

ML_R_PlotSketch(ax=ax, N=50000)
plt.plot(s, R, marker='.', linestyle='none', color='grey', markersize=8, clip_on=False)
plt.errorbar(mean_s, mean_R, xerr=np.std(s), yerr=np.std(R), color='gold', zorder=10, capsize=2)
plt.show()
print(mean_R, np.std(R))
plt.tight_layout()
fig.savefig("../reports/ML_HitAndMiss_S2results_1000fits.nc"+"figure_R_sketch.pdf", 
            dpi=300, bbox_inches="tight")
```

```python
fig, ax = plt.subplots(figsize=(1.25,1.25), dpi=300)

mean_s_S1 = np.mean(s_S1)
mean_R_S1 = np.mean(R_S1)

mean_s_S2 = np.mean(s_S2)
mean_R_S2 = np.mean(R_S2)

ML_R_PlotSketch(ax=ax, N=50000, linestyle='--', color='gold')
plt.scatter(s_S1, R_S1, s=10, c='grey', edgecolor='none', alpha=1, zorder=8, clip_on=False)
plt.errorbar(mean_s_S1, mean_R_S1, xerr=np.std(s_S1), yerr=np.std(R_S1), color='gold', zorder=10, capsize=2)

ML_R_PlotSketch(ax=ax, N=6000, linestyle='--', color='pink')
plt.scatter(s_S2, R_S2, s=10, c='grey', edgecolor='none', alpha=1, zorder=8, clip_on=False)
plt.errorbar(mean_s_S2, mean_R_S2, xerr=np.std(s_S2), yerr=np.std(R_S2), color='pink', zorder=10, capsize=2)

plt.show()
plt.tight_layout()
fig.savefig("../reports/ML_HitAndMiss_S1andS2results_1000fits.nc"+"figure_R_sketch.pdf", 
            dpi=300, bbox_inches="tight")
```
