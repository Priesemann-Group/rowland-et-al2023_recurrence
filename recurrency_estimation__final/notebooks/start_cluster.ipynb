```python
from dask_jobqueue import SGECluster
```

```python
#cluster.adapt(minimum=0, maximum=maximal_number_of_nodes*processes_per_node, interval='60s', wait_count=10)
#cluster.scale(maximal_number_of_nodes*processes_per_node)
```

```python
cluster_name = "zal"

maximal_number_of_nodes = 24
processes_per_node = 2

cluster = SGECluster(
    # we have 32 cores per node and 192 GB RAM
    cores=processes_per_node,
    memory='192GB',
    # only launch 1 process per node, with 32 threads /this seems to be our
    # HPC convention. For dask this means 1 job == 1 worker
    processes=processes_per_node,
    walltime="1500000",
    # use `job_extra` instead of `queue`. this gives `-pe` instead of `-q` in the
    # job script which we need to get the parallel environment, match number cores
    job_extra=["-pe mvapich2-"+cluster_name+" 32"],
    # use fast directories for logs etc. beware scratch01 on sohrab is not scratch on sam
    # you have to create the direcotires manually, beforehand.
    log_directory="/scratch04.local/loidolt/dask/logs/",
    local_directory="/scratch04.local/loidolt/dask/scratch/",
    # we want to use the fast `ib0` but I have not managed yet.
    interface="ib0",
    # do not start workers automatically. for us, n_wokers corresponds to number of nodes
    n_workers=0,
    # if you (like me) have some hackery scripts each worker needs to run at launch
    # you can use `extra`
    death_timeout=60*30,
    extra=[
        '--preload cluster_worker_preload.py'
    ],
    silence_logs='debug',
    scheduler_options={'port':46466}
)

#cluster.adapt(minimum=0, maximum=maximal_number_of_nodes*processes_per_node, interval='30s', wait_count=5)
#cluster.scale(maximal_number_of_nodes*processes_per_node//2)
cluster.scale(maximal_number_of_nodes*processes_per_node)
```
